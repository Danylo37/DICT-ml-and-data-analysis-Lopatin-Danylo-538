# 1. Числові характеристики ознак

```python
import pandas as pd

file_path = 'iris.csv'
df = pd.read_csv(file_path)

print("\nStatistics of the dataset:")
df.describe()
```

# 2. Збалансованість класів

```python
import matplotlib.pyplot as plt
import os

os.makedirs('results', exist_ok=True)

species_counts = df['Species'].value_counts()

plt.figure(figsize=(8, 6))
species_counts.plot(kind='bar', color=['lightcoral', 'lightgreen', 'skyblue'])
plt.title('Distribution of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('results/species_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
```

# 3. Візуалізация

## Sepal Length vs Sepal Width

```python
plt.figure(figsize=(10, 7))
for name, group in df.groupby('Species'):
    plt.scatter(group['Sepal.Length'], group['Sepal.Width'], s=100, label=name)
plt.title('Sepal Length vs. Sepal Width by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.grid(alpha=0.3)
plt.legend(title='Species')
plt.savefig('results/sepal_length_vs_sepal_width.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Petal Length vs Petal Width

```python
plt.figure(figsize=(10, 7))
for name, group in df.groupby('Species'):
    plt.scatter(group['Petal.Length'], group['Petal.Width'], s=100, label=name)
plt.title('Petal Length vs. Petal Width by Species')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.grid(alpha=0.3)
plt.legend(title='Species')
plt.savefig('results/petal_length_vs_petal_width.png', dpi=300, bbox_inches='tight')
plt.show()
```
